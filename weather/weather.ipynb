```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dailyclimate.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
df['year'] = pd.to_datetime(df['Date']).dt.year
df['month'] = pd.to_datetime(df['Date']).dt.month
df['day'] = pd.to_datetime(df['Date']).dt.day

# convert int_type data into float_type
df['year'] = df['year'].astype(float) 
df['month'] = df['month'].astype(float) 
df['day'] = df['day'].astype(float) 
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Location'] = encoder.fit_transform(df['District'])
```

```python
place = df[['District', 'year', 'month', 'day']]
place
```

```python
df = df.drop('District', axis=1)
df
```

```python
got_para = ['Longitude', 'Latitude', 'Temp_2m', 'MaxTemp_2m', 'MinTemp_2m', 'Pressure', 'Humidity_2m', 'WindSpeed_10m']
```

```python
# 1kPa(pressure) = 10hPa(OpenWeather)
# from the city data given by user convert that into latitude and longitude
```

```python
# 1kPa(pressure) = 10hPa(OpenWeather)
df['pressure'] = df['Pressure']
df = df.assign()
df.loc[:,'Pressure'] = df['pressure']*100
df = df.drop('pressure', axis=1)
df
```

```python
use_df = df[got_para]
use_df
```

```python
use_df.info()
```

```python
from sklearn.model_selection import train_test_split
x = use_df.drop(['Temp_2m'], axis=1)
y = use_df['Temp_2m']

# Split into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
model.fit(x_train, y_train)

# Make predictions
y_pred = model.predict(x_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)
print(f"R-squared Score: {r2}")
```

```python
import key
import requests
import os

API_key = key.API_KEY
def weather():
    print("\n Weather now")
    try:
        city = input("Please enter the city name: ")
        request_url = f'https://api.openweathermap.org/data/2.5/weather?appid={API_key}&q={city}&units=metric'
        weather_data = requests.get(request_url).json()
#         print(weather_data)
        print(f'\n Current temperature for {weather_data["name"]}: ')
        new_data = {
                    'Longitude' : [weather_data['coord']['lon']],
                    'Latitude' : [weather_data['coord']['lat']],
                    'MaxTemp_2m': [weather_data['main']['temp_max']], 
                    'MinTemp_2m': [weather_data['main']['temp_min']], 
                    'Pressure': [weather_data['main']['pressure']],
                    'Humidity_2m': [weather_data['main']['humidity']], 
                    'WindSpeed_10m': [weather_data['wind']['speed']]
                    }
        new_data_df = pd.DataFrame(new_data)
        predicted_temp = model.predict(new_data_df)
        return f'The temperature is {predicted_temp}'
#         print(f'The temperature is {weather_data ["main"]["temp"]} ,pressure is {weather_data ["main"]["pressure"]} and humidity is {weather_data ["main"]["temp"]}.')
    except KeyError:
        return "Please enter a valid name for city."
```
