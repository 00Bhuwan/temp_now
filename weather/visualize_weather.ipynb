```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('dailyclimate.csv')
```

```python
# Date and District arein object datatype and unnamed:0 is int data type 
data.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index("Date", inplace=True)
```

```python
data.isnull().sum()
```

```python
data.describe().T
```

```python
data.duplicated().sum()
```

```python
data.corr(numeric_only=True)['Temp_2m']
```

```python
plt.figure(figsize=(14,6))
plt.plot(data['Precip'], label='Precipitation', color='blue')
plt.plot(data['Temp_2m'], label='Temperature', color='orange')
plt.title('Precipitation and Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='Latitude', y='Longitude', data=data, hue='Precip', palette='viridis', size='Temp_2m', sizes=(20,200))
plt.title('Distrubution of precipitation and temperature')
plt.show()
```

```python
corr_data = data.drop('District', axis=1)
correction_matrix = corr_data.corr()

plt.figure(figsize=(22,15))
sns.heatmap(correction_matrix, annot=True, cmap='coolwarm', linewidth=.1)
plt.show()
```

```python
sns.histplot(data[['Temp_2m', 'MaxTemp_2m', 'MinTemp_2m']], kde=True, bins=30)
plt.xlabel('Temp(degree celcius)')
plt.show()
```

```python
sns.histplot(data[['Humidity_2m', 'RH_2m']], kde=True, bins=30)
plt.title('Humidity')
plt.show()
```

```python
sns.kdeplot(data=data[['WindSpeed_10m', 'WindSpeed_50m']])
plt.title('Distribution of wind between 10m and 50m')
plt.show()
```

```python
key_para=['Precip', 'Temp_2m', 'Pressure', 'Humidity_2m', 'WindSpeed_10m']
sns.pairplot(data[key_para])
plt.show()
```
